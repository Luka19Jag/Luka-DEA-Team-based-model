## Third constraint

Active players in third constraint are players who are limited to have better efficient index, because of very poor efficiency of their's teammates. This means that they would be more efficient if there weren't new third restrictions, but with this condiotion they have problem to allocate resources in efficient way.

```python
# Third constraint

third_con_07 = pd.read_csv('D:\Luka Sve\Strucni radovi\DEA\Rezultati - tabele\Rezultati modificated - team DEA\gama = 0.7/third_constraints_0.7.csv')
third_con_07.rename(columns={'Unnamed: 0' : 'player_name'}, inplace=True)
third_con_07.set_index('player_name', inplace=True)
third_con_07 = round(pd.merge(third_con_07, dic_eff_07[['Efficiency','club', 'position']], on='player_name', how='inner'), 3)
third_con_07
```

```python
# We seek players which slack value is very close to 0
# All of these players are active constraints

act_third_con_07 = third_con_07[(third_con_07['Slack'] < 0.002)]
act_third_con_07
```

```python
# These players are in active constraints, but still efficient 
# These are importanat to analyse
# This is only for Juventus from the efficient teams
# Haller, and Klaassen, from Ajax, are also close to be active third contraints -> put value of 0.01
    
act_third_con_eff_07 = third_con_07[(third_con_07['Slack'] < 0.002) & (third_con_07['Efficiency'] == 1)]
act_third_con_eff_07
```

```python
act_third_con_eff2_07 = third_con_07[(third_con_07['Slack'] < 0.01) & (third_con_07['Efficiency'] == 1)]
act_third_con_eff2_07
```

```python
act_third_con_07_clubs = act_third_con_07.groupby('club').count()[['Slack']]
act_third_con_07_clubs
```
