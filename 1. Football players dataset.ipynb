```python
# method for vizualization dataframes

def color_false_red(val):
    if val == False:
        color = 'red'
    else:
        color = 'green'
    return 'color: %s' % color
```

```python
# method for vizualization dataframes

def highlight(x):
    c1 = 'color: green'
    c2 = 'color: red'
    c3 = '' 
    #m = x.eq(x.iloc[:, 0], axis=0)
    m = x != True
    n = x != False
    df1 = pd.DataFrame(c3, index=x.index, columns=x.columns)
    #add color red by False values 
    df1 = df1.where(m, c1)
    df1 = df1.where(n, c2)
    return df1
```

# 2. Filter method - F-test

These are output variables:
y1='goals'
y2='assists'
y3='balls_recoverd'

```python
y1 = stats['goals']
y2 = stats['assists']
y3 = stats['balls_recoverd']
X = stats.drop(output_att + categorial_att, axis=1)
```

It will be done F-test for every output variable, and it will be calculated 10 the most informative attributes.

## 2.1. F-test : k-best=10 for 'goals'

```python
selection10_goals = SelectKBest(f_regression, k=10)
selection10_goals.fit(X, y1)
X_filter_k10_goals = X.loc[:, selection10_goals.get_support()]
```

```python
dfk10Atr_goals = pd.DataFrame({'Column' : X.columns.tolist(),
                   'Value True/False, k=10, goals' : selection10_goals.get_support().tolist()})
dfk10Atr_goals = dfk10Atr_goals.set_index('Column')
dfk10Atr_goals.style.applymap(color_false_red)
```

## 2.2. F-test : k-best=10 for 'assists'

```python
selection10_assists = SelectKBest(f_regression, k=10)
selection10_assists.fit(X, y2)
X_filter_k10_assists = X.loc[:, selection10_assists.get_support()]
```

```python
dfk10Atr_assists = pd.DataFrame({'Column' : X.columns.tolist(),
                   'Value True/False, k=10, assists' : selection10_assists.get_support().tolist()})
dfk10Atr_assists = dfk10Atr_assists.set_index('Column')
dfk10Atr_assists.style.applymap(color_false_red)
```

## 2.3 F-test : k-best=10 for 'balls_recoverd'

```python
selection10_balls_recoverd = SelectKBest(f_regression, k=10)
selection10_balls_recoverd.fit(X, y3)
X_filter_k10_balls_recoverd = X.loc[:, selection10_balls_recoverd.get_support()]
```

```python
dfk10Atr_balls_recoverd = pd.DataFrame({'Column' : X.columns.tolist(),
                   'Value True/False, k=10, balls_recoverd' : selection10_balls_recoverd.get_support().tolist()})
dfk10Atr_balls_recoverd = dfk10Atr_balls_recoverd.set_index('Column')
dfk10Atr_balls_recoverd.style.applymap(color_false_red)
```
